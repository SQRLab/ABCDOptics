```python
from opticalSystemClass import OpticalSystem
import matplotlib.pyplot as plt
```

```python
"""
    This script is used to test the beamFunctions.py script. We are going to generate a beam with some parameters and then propagate it
    through an optical system and then plot it. Currently the radius of curvature needs to be added to the beamFunctions.py module.
"""

# Parameters of the beam
wavelength = 422e-9 #422 nm
waist = 500e-6 #beam waist (um)
z = 0 #initial position of the beam
theta = 0 #initial angle of the beam (rad)

system = OpticalSystem(wavelength, waist, z, theta) #generates initial beam

# Generate optics in the path of the beam
system.freeSpace(.15) #free space of 0.1 m
system.lens(0.15) #lens with focal length of 0.15 m
system.freeSpace(.15) #free space of 0.1 m

system.propagate_beam() #propagates the beam through the optical system

print(f"The intial waist of the beam is {waist*1e6} um, initial angle is {theta} rad and position is {z} m")
print(f"The final waist of the beam is {system.waist_out*1e6} um")

print(system.q_out)
```
